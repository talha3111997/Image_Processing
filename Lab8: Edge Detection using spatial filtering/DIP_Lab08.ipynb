```
import cv2
from PIL import Image
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
#------Lab 08 Task 1 ------------
img = cv2.imread('building.tif',0)
cv2_imshow(img)
```

```
def applySobelFilter(img):
  sobelX = np.array(([-1, 0, 1],
                  [-2, 0, 2],
                  [-1,0, 1]))
  sobelY = np.array(([-1, -2, -1],
                  [0, 0, 0],
                  [1, 2, 1]))
  imgSobelX = cv2.filter2D(img, -1, sobelX)
  imgSobelY = cv2.filter2D(img, -1, sobelY)
  imgSobelFinal = abs(imgSobelX) + abs(imgSobelY)
  return imgSobelFinal;
```

```
imgSobelFinal = applySobelFilter(img)
```

```

cv2_imshow(imgSobelFinal)
```

```
def applyPrewittFilter(img):
  prewittX = np.array(([-1, 0, 1],
                  [-1, 0, 1],
                  [-1,0, 1]))
  prewittY = np.array(([-1, -1, -1],
                  [0, 0, 0],
                  [1, 1, 1]))
  imgPrewittX = cv2.filter2D(img, -1, prewittX)
  imgPrewittY = cv2.filter2D(img, -1, prewittY)
  imgPrewittFinal= abs(imgPrewittX) + abs(imgPrewittY)
  return imgPrewittFinal
```

```
imgPrewittFinal = applyPrewittFilter(img)
```

```
cv2_imshow(imgPrewittFinal)
```

```
# smoothing the original image

smoothImg = cv2.GaussianBlur(img,(5,5),0)
cv2_imshow(smoothImg)
```

```

def applyMask(img , threshold):
  mask=np.zeros((img.shape))
  for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if(img[row,col]>threshold):
          mask[row][col]=255
        else:
          mask[row][col]=0
  return mask;
```

```
smoothImgSober = applySobelFilter(smoothImg)
plt.imshow(smoothImgSober, cmap='gray')
```

```
#  Sober filter
mask=applyMask(smoothImgSober, (0.50*255))
plt.imshow(mask, cmap='gray')
```

```
# Prewitt filter
smoothImgPrewitt = applyPrewittFilter(smoothImg)
mask=applyMask(smoothImgPrewitt, (0.50*255))
plt.imshow(mask, cmap='gray')
```

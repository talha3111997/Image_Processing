```
import cv2
from PIL import Image
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
import math
import skimage.filters 
from scipy.ndimage import filters
```

```
img = cv2.imread("pears.png", 0)
cv2_imshow(img)
```

```
type(np.asarray(img))
```

```
# 255 is used because the below function gives the images in [0 , 1] range i.e. in normalize form
gaussianNoise = skimage.util.random_noise(img,mode='gaussian')*255
cv2_imshow(gaussianNoise)
```

```
np.max(gaussianNoise)
```

```
poissonNoise = skimage.util.random_noise(img,mode='poisson')* 255
cv2_imshow(poissonNoise)
```

```
saltpepperNoise = (skimage.util.random_noise(img,mode='s&p')*255)
cv2_imshow(saltpepperNoise)
```

```
speckleNoise = (skimage.util.random_noise(img,mode='speckle')*255)
cv2_imshow(speckleNoise)
```

```
saltNoise = (skimage.util.random_noise(img,mode='salt')*255)
cv2_imshow(saltNoise)
```

```
pepperNoise = (skimage.util.random_noise(img,mode='pepper')*255)
cv2_imshow(pepperNoise)
```

```
# ------- Task 2 ---------------------------------------------
kernel = np.array([ 1, 1, 1, 1, 1 , 1, 1, 1, 1], dtype="float32").reshape(3, 3)/9
```

```
kernel
```

```
MeanImgGaussian =cv2.filter2D(gaussianNoise , -1 ,  kernel)
cv2_imshow(MeanImgGaussian) 
```

```
MeanImgPoisson =cv2.filter2D(poissonNoise, -1 ,  kernel)
cv2_imshow(MeanImgPoisson) 
```

```
salt_pepper_min_filter_img = filters.minimum_filter(saltpepperNoise, (3,3))
cv2_imshow(salt_pepper_min_filter_img)
```

```
salt_pepper_max_filter_img = filters.maximum_filter(saltpepperNoise, (3,3))
cv2_imshow(salt_pepper_max_filter_img)
```

```
salt_pepper_median_filter_img = filters.median_filter(saltpepperNoise, (3,3))
cv2_imshow(salt_pepper_median_filter_img)
```

```
# Geometric mean
cv2_imshow(skimage.filters.rank.geometric_mean(speckleNoise/255, kernel))
```

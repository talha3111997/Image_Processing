```
import cv2
from PIL import Image
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```
img2= cv2.imread('coins.jpg',0)
```

```
img2.shape
```

```
cv2_imshow(img2)
```

```
intVal = np.zeros((256), dtype='uint32')
```

```
mask= np.zeros((img2.shape[0],img2.shape[1]), dtype='uint32')
```

```
for row in range(img2.shape[0]):
  for col in range(img2.shape[1]):
    intVal[img2[row][col]]+=1
```

```
intVal
```

```
plt.plot(intVal)
```

```
threshold=100
for row in range(img2.shape[0]):
  for col in range(img2.shape[1]):
    if(img2[row][col]>100):
      mask[row][col]=255
    else:
      mask[row][col]=0
```

```
cv2_imshow(mask)
```

```
wantedImg = np.bitwise_and(img2, mask)
```

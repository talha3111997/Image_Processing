```
img2=cv2.imread("aerial.tif", 0)
```

```
img2
```

```
cv2_imshow(img2)
```

```
hist=cv2.calcHist([img2],[0],None,[256],[0,256])
plt.plot(hist)
```

```
gamaImg = np.zeros((img2.shape[0],img2.shape[1]), dtype='uint8')
```

```
def powerLaw(x):
  c=255
  for row in range(img2.shape[0]):
    for col in range(img2.shape[1]):
      gamaImg[row][col]=(c*((img2[row][col]/255)**x))
      #print(c*(img2[row][col]**x)%255)
      #print(gamaImg[row][col])
      
  cv2_imshow(gamaImg)

interact(powerLaw, x=widgets.FloatSlider(min=0.0, max=15.0, step=0.1, value=0.5));

```

```
hist=cv2.calcHist([gamaImg],[0],None,[256],[0,256])
plt.plot(hist)
```

```
import numpy as np
from PIL import Image
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```
img = cv2.imread('imgLara.png',0)
# Without converting into binary img
cv2_imshow(img)
```

```
# Structuring Element
SE = np.ones((3,3), np.uint8)
```

```
# Erosion with different iterations, here erosion is increasing the black pixels
img_erosion = cv2.erode(img, SE, iterations=1) 

cv2_imshow(img_erosion)
```

```
img_erosion = cv2.erode(img, SE, iterations=5) 

cv2_imshow(img_erosion)
```

```
# Dilation experiment with different iterations, here dilation means increasing the white pixels
img_dilation = cv2.dilate(img, SE, iterations=1) 

cv2_imshow(img_dilation)
```

```
img_dilation = cv2.dilate(img, SE, iterations=2) 

cv2_imshow(img_dilation)
```

```
img_dilation = cv2.dilate(img, SE, iterations=5) 

cv2_imshow(img_dilation)
```

```
#----------------------------------- Task 2 -----------------------------
img2 = cv2.imread('imgUrdu.png',0)
ret, binImg2 = cv2.threshold(img2,127,255,cv2.THRESH_BINARY)
cv2_imshow(binImg2)
```

```

SE2=cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

SE2
```

```
img2_dilate = cv2.dilate(binImg2, SE2, iterations=4) 

cv2_imshow(img2_dilate)
```

```
img2_erosion = cv2.erode(img2_dilate, SE, iterations=8) 

cv2_imshow(img2_erosion)

# Tasks Finished
```

```
# Extra just for fun, (difference between dilation and erosion of an image.)

cv2_imshow(cv2.bitwise_not(img2_erosion)-cv2.bitwise_not(img2_dilate))
```
